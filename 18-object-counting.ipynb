```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
# duong dan anh
my_dir = 'C:/Users/Admin/Documents/CVs/img/'
my_path = my_dir + "5.jpg"

# doc anh
img = Image.open(my_path)

#chuyen sang array anh xam
img_gray = img.convert('L')

img_gray_arr = np.array(img_gray)

#chuyen sang anh nhi phan
img_bi = 1*(img_gray_arr<225) #bat ki diem nao < 200 => 1, nguoc lai se la 0

# label anh
img_labeled, num_objs = ndimage.label(img_bi)

# tao luoi subplot voi 3 dong 1 cot
fig, axs = plt.subplots(3,1,figsize=(50,50))

plt.subplot(3,1,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Orginal gray")

plt.subplot(3,1,2)
plt.imshow(img_bi, cmap='gray')
plt.title("Binary image")

plt.subplot(3,1,3)
plt.imshow(img_labeled, cmap='gray')
plt.title(f"labeled image {num_objs}")

plt.show()
```

```python
# Morphology (hình thái học)
#Erosion: xem xét 9 pixel xung quanh
# nếu tất cả là 1 (trong ảnh binary, thì điểm đó là 1)
# dùng để thủ nhỏ obj size trong ảnh nhị phân, loại bỏ pixel nhiễu
# Opening = (Erosion -> Dilation): thường dùng loại bỏ nhiễu và thu nhỏ obj trong ảnh
# Ứng dụng: Loại bỏ các điểm nhiễu nhỏ nhưng giữ nguyên hình dạng đối tượng lớn.

# Closing = (Dilation -> Erosion)
# Ứng dụng: Khắc phục lỗi ngắt đoạn của đối tượng trong ảnh nhị phân.
# thuc hien opening de loai bo nhieu va noi cac doi tuong
img_opening = ndimage.binary_closing(img_bi, structure=np.ones((15,5)))

#thuc hien label lai
img_labeled, num_objs = ndimage.label(img_opening)

# tao luoi subplot voi 4 dong 1 cot
fig, axs = plt.subplots(4,1,figsize=(50,50))

plt.subplot(4,1,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Orginal gray")

plt.subplot(4,1,2)
plt.imshow(img_bi, cmap='gray')
plt.title("Binary image")

plt.subplot(4,1,3)
plt.imshow(img_opening, cmap='gray')
plt.title("Anh sau openning")

plt.subplot(4,1,4)
plt.imshow(img_labeled, cmap='gray')
plt.title(f"labeled image {num_objs}")

plt.show()
```
